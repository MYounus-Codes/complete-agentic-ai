## **Web Search Agent Using GROQ built-in feature in `groq/compound` model**

```python
from langchain.agents import create_agent
from langchain.tools import tool
import os

os.environ["GROQ_API_KEY"] = "YOUR_GROQ_API_KEY"

from langchain_groq import ChatGroq

llm = ChatGroq(model="groq/compound")   

# 3. Updated System Prompt (Forces the Agent to resolve countries to capital/major cities)
agent = create_agent(
    model=llm,
    system_prompt=(
        "You are a helpful web search assistant."
    ),
)

response = agent.invoke({"messages":[{"role": "user", "content": "Who wins the fifa world cup 2026 first match?"}]})
print(response["messages"][-1].content)
```
