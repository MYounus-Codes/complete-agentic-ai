#### **MODEL INTEGRATION WITH `OPENAI, GOOGLE GEMINI, GORQ, ANTHROPIC CLAUDE`**
----
#### **There are two ways to integrate the `model` with `langchain`:**
- **init_chat_models**
- **specific provider package**, e.g: `langchain-groq` `langchain-openai` `langchain-google-genai` `langchain-anthropic`

### **Use any model with `langchain.chat_models`**

```python
## Use Any Model
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY") ## Just make sure to set your GROQ_API_KEY in your environment variables before running this code

model = init_chat_model("groq:qwen/qwen3-32b") ## Just change the model name here to use a different model
response = model.invoke("Why man do coding????") 
print(response.content)
```

### **Use models with there specific packagaes**

```python
## GROQ

## uv add langchain-groq | pip install langchain-groq

import os
from langchain_groq import ChatGroq

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0,
    max_tokens=None,
    reasoning_format="parsed",
    timeout=None,
    max_retries=2,
    # other params...
)

response = llm.invoke("What is the capital of France?")
print(response)
```

```python
## GOOGLE GEMINI

## uv add langchain-google-genai | pip install langchain-google-genai   

import os
from langchain_google_genai import ChatGoogleGenerativeAI

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    # other params...
)

response = llm.invoke("Who won first match of FIFA World Cup 2026?")
print(response) 
```

```python
## OPENAI 

## uv add langchain-openai	| pip install langchain-openai

import os
from langchain_openai import ChatOpenAI

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

llm = ChatOpenAI(
    model="gpt-4o",
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)

response = llm.invoke("What is new feature released in Codex? short answer only")
print(response)
```

```python
## Anthropic Claude

## uv add langchain-anthropic | pip install langchain-anthropic

import os
from langchain_anthropic import ChatAnthropic

os.environ["ANTHROPIC_API_KEY"] = os.getenv("ANTHROPIC_API_KEY")

llm = ChatAnthropic(
    model="claude-2",
    temperature=0.6,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)

response = llm.invoke("What is new feature released in Claude sonnet? short answer only")
print(response)
```
