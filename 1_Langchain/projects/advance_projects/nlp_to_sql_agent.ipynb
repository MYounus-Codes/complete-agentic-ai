```python
from langchain_community.utilities import SQLDatabase
from langchain_community.agent_toolkits import create_sql_agent
from langchain_groq import ChatGroq   # or ChatGroq
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

# 1. Connect to MySQL
db = SQLDatabase.from_uri(
    "mysql+pymysql://root:1234@localhost:3306/my_database" # replace with your MySQL connection string
)

# 2. Load LLM
llm = ChatGroq(model="qwen/qwen3-32b", temperature=0)

# 3. Create SQL Agent
agent = create_sql_agent(
    llm=llm,
    db=db,
    verbose=True
)

# 4. Ask question
response = agent.invoke(
    {"input": "show me users who have salary higher than 70000 and are males and order by salary desc limit 5"}
)
print(response)
```
