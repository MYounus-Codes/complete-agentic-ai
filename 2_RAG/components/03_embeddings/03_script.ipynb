## **Embeddings — The Mathematical Heart of RAG**

#### **What is an Embedding?**

***An embedding is a numerical representation (a list of numbers) of a piece of text that captures its meaning.***
---
#### **In simple terms:**

***`Embedding = Text → Numbers that preserve semantic meaning.`***

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
    model_name="BAAI/bge-small-en-v1.5"
)
```

```python
## Embed a query

text = "My name is Muhammad Younus"

vector = embedding_model.embed_query(text)

print(f"Dimensions: {len(vector)}")
print(vector[:10])
```

```python
## Embed a document
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 1. Load the document
loader = PyMuPDFLoader("AKU_EB_Physics_Chapter1_Notes.pdf")
documents = loader.load()

# 2. Split the document into chunks
splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=110
)
chunks = splitter.split_documents(documents)

# 3. Extract the text content from each chunk
chunk_texts = [chunk.page_content for chunk in chunks]

# 4. Embed the text chunks
vectors = embedding_model.embed_documents(chunk_texts)

# 5. Verify the results
print(f"Total Chunks: {len(vectors)}")
print(f"Dimensions: {len(vectors[0])}")
# print(f"First 10 values of first chunk: {vectors[0][:10]}")
print(vectors)
```
