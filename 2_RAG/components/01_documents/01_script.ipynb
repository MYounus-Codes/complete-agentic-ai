## **Documents and Knowledge Sources in RAG**

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("doc.txt")
documents = loader.load()
print(documents)
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_classic.schema import Document

loader = PyMuPDFLoader("AKU_EB_Physics_Chapter1_Notes.pdf")
documents = loader.load()
print(documents)
print(len(documents))
print(documents[0].metadata)
print(documents[0].page_content)
```

```python
from langchain_community.document_loaders import UnstructuredURLLoader
loader = UnstructuredURLLoader(urls=["https://en.wikipedia.org/wiki/Artificial_general_intelligence"], headers={"User-Agent": "Mozilla/5.0"})
documents = loader.load()
print(documents)
```
