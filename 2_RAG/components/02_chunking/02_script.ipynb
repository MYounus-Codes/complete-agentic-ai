## **Chunking (Text Splitting) in RAG**

##### **What is Chunking?**
- ***Chunking is the process of breaking large documents into smaller pieces called chunks.***

#### **Overlap**

**Chunk 1:**
- Sentence 1
- Sentence 2
- Sentence 3

**Chunk 2:**
- Sentence 3
- Sentence 4
- Sentence 5

#### **Basic Example**

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text = """
Python is a programming language.
It is widely used in AI.
Python is easy to learn.
"""

splitter = RecursiveCharacterTextSplitter(
    chunk_size=50,
    chunk_overlap=10
)

chunks = splitter.split_text(text)

print(chunks)
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = PyMuPDFLoader("AKU_EB_Physics_Chapter1_Notes.pdf")
documents = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size=600, 
    chunk_overlap=110
)
chunks = splitter.split_documents(documents)

print(len(chunks))
# print(chunks)

for  i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}:")
    print(chunk.page_content)
    print("\n---\n")
```
